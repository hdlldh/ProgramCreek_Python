# 1. Two Sum

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        hashmap = {}
        for i in range(len(nums)):
            if nums[i] not in hashmap:
                hashmap[target-nums[i]] = i
            else:
                return [hashmap[nums[i]],i]
        return []
```

# 3. Longest Substring Without Repeating Characters

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        hmap = {}
        ans = 0
        for i in range(len(s)):
            if s[i] not in hmap:
                hmap[s[i]] = i
            else:
                prev = hmap[s[i]]
                ans = max(ans,len(hmap))
                hmap = {}
                for j in range(prev+1,i+1):
                    hmap[s[j]] = j
        ans = max(ans,len(hmap))
        return ans
```

# 7. Reverse Integer

```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        
        isPos = x>=0
        x = abs(x)
        y = 0
        while x>0:
            y = y*10+ (x%10)
            x = x//10
        y = y if isPos else -y
        if y>=2**31 or y< -2**31: return 0
        return y 
```

# 11. Container With Most Water

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        n = len(height)
        ans = 0
        i = 0
        j = n-1
        while i<j:
            area = min(height[i],height[j])*(j-i)
            ans = max(ans, area)
            if height[i]< height[j]:
                i+=1
            else:
                j-=1
        return ans
```

# 15. 3Sum

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        nums.sort()
        n = len(nums)
        ans = []
        for i in range(n):
            if i!=0 and nums[i]==nums[i-1]:continue
            j=i+1
            k=n-1
            while j<k:
                s = nums[i]+nums[j]+nums[k]
                if s == 0:
                    ans.append([nums[i],nums[j],nums[k]])
                    j+=1
                    while (j<k and nums[j-1]==nums[j]): 
                        j+=1
                elif s<0:
                    j+=1
                else:
                    k-=1
        return ans
                    
```

```python
class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        nums.sort()
        n = len(nums)
        ans = []
        for i in range(n-1):
            if i!=0 and nums[i]==nums[i-1]:continue
            for j in range(i+1,n):
                if j!=i+1 and nums[j]==nums[j-1]:continue
                k=j+1
                l=n-1
                while k<l:
                    s=nums[i]+nums[j]+nums[k]+nums[l]
                    if s==target: 
                        ans.append([nums[i],nums[j],nums[k],nums[l]])
                        k+=1
                        while(k<l and nums[k-1]==nums[k]):k+=1
                    elif s<target: 
                        k+=1
                    else:
                        l-=1
        return ans
            
```

# 20. Valid Parentheses

```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if not s: return True
        stack = []
        left = ["(","[","{"]
        right = [")","]","}"]
        for i in range(len(s)):
            if s[i] in left:
                stack.append(s[i])
            else:
                if len(stack) > 0:
                    temp = stack.pop()
                    if left.index(temp)!=right.index(s[i]): return False
                else:
                    return False
        if len(stack)!= 0: return False
        return True
```

# 21. Merge Two Sorted Lists

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        l = ListNode(None)
        p = l
        while l1 or l2:
            if l1 and l2:
                if l1.val<= l2.val:
                    p.next = l1
                    p = p.next
                    l1 = l1.next
                else:
                    p.next = l2
                    p = p.next
                    l2 = l2.next
            elif l1:
                p.next = l1
                p = p.next
                l1 = l1.next
            else:
                p.next = l2
                p = p.next
                l2 = l2.next
        return l.next
```

# 29. Divide Two Integers

```python
class Solution(object):
    def divide(self, dividend, divisor):
        """
        :type dividend: int
        :type divisor: int
        :rtype: int
        """
        if dividend < 0:
            dividend = -dividend
            pos1 = False
        else:
            pos1 = True
        if divisor <0:
            divisor = -divisor
            pos2 = False
        else:
            pos2 = True
        
        if dividend < divisor: return 0
        ans = 0
        while dividend>=divisor:
            div = divisor
            res = 1
            while div*2 < dividend:
                div *=2
                res *=2
            dividend -= div
            ans += res
        
        if pos1!=pos2:
            return max(-2**31,-ans)
        else:
            return min(2**31-1,ans)
```

# 33. Search in Rotated Sorted Array

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        n = len(nums)
        return self.helper(nums,0,n-1, target)
        
    def helper(self, nums, l, r, target):
        if l>r: return -1
        mid = (l+r)//2
        if nums[mid] == target:
            return mid
        if nums[l] <= nums[mid]:
            if nums[l] <= target <= nums[mid]:
                return self.helper(nums,l,mid-1, target)
            else:
                return self.helper(nums,mid+1,r, target)
        elif nums[mid]<= nums[r]:
            if nums[mid] <= target <= nums[r]:
                return self.helper(nums,mid+1,r,target)
            else:
                return self.helper(nums,l,mid-1,target)
        return -1
            
        
```

# 34. Find First and Last Position of Element in Sorted Array

```python
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        low = 0
        high = len(nums)-1
        if low>high: return [-1,-1]
        while low<=high:
            mid = (low+high)//2
            if nums[mid] == target:
                break
            elif nums[mid]< target:
                low = mid+1
            else:
                high = mid-1
        if nums[mid]== target:
            start = end = mid
            while start>=0 and nums[start]==nums[mid]:
                start -= 1
            while end<len(nums) and nums[end]==nums[mid]:
                end += 1
            return [start+1,end-1]
        return [-1,-1]
        
```

# 35. Search Insert Position

```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        if not nums: return 0
        if target <= nums[0]: return 0
        low = 0
        high = len(nums)
        
        while (high -low)>1:
            mid = (low+high)//2
            if nums[mid] == target:
                return mid
            elif nums[mid]< target:
                low = mid
            else:
                high = mid
        return high
```

# 38. Count and Say

```python
class Solution(object):
    def countAndSay(self, n):
        """
        :type n: int
        :rtype: str
        """
        if n ==1: return '1'
        
        start = '1'
        for _ in range(n-1):
            ans = self.helper(start)
            start = ans
        return start
    
    def helper(self,s):
        start = s[0]
        count = 1
        out = ''
        for i in range(1,len(s)):
            if s[i] != start:
                out+= str(count)+ start
                start = s[i]
                count = 1
            else:
                count += 1
        if count: out+= str(count)+start
        return out 
                
```

# 39. Combination Sum

```python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        candidates.sort()
        self.ans = []
        self.helper(candidates,target,0,[])
        return self.ans
        
    def helper(self,candidates, target, idx, path):
        if target == 0:
            self.ans.append(path)
            return
        
        if target < 0:
            return
        
        for i in range(idx,len(candidates)):
            if candidates[i] > target: break
            self.helper(candidates,target-candidates[i],i,path+[candidates[i]])
            
```

# 43. Multiply Strings

```python
class Solution(object):
    def multiply(self, num1, num2):
        """
        :type num1: str
        :type num2: str
        :rtype: str
        """
        if not num1 or not num2: return '0'
        if num1=='0' or num2=='0': return '0'
        l1 = [int(n) for n in num1]
        l1.reverse()
        l2 = [int(n) for n in num2]
        l2.reverse()
        ans = [0]*(len(l1)+len(l2))
        for i1 in range(len(l1)):
            for i2 in range(len(l2)):
                product = l1[i1]*l2[i2]
                p1 = i1+i2
                p2 = i1+i2+1
                sum = ans[p1]+product
                ans[p1] = sum%10
                ans[p2] += sum//10
        ans.reverse()
        ans = ''.join([str(d) for d in ans])
        return ans.lstrip('0')
                
                
```

# 46. Permutations

```python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        if not nums: return []
        ans = []
        n = len(nums)
        for _ in range(n):
            if not ans:
                for i in nums:
                    ans.append([i])
            else:
                tmp = []
                for sub in ans:
                    for i in nums:
                        if i not in sub:
                            new = [element for element in sub] + [i]
                            tmp.append(new)
                ans = tmp
        return ans
```

# 47. Permutations II

```python
from collections import Counter
class Solution(object):
    def permuteUnique(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        if not nums: return []
        ans = []
        n = len(nums)
        hmap = Counter(nums)
        for _ in range(n):
            if not ans:
                for i in hmap:
                    ans.append([i])
            else:
                next_ans = []
                for sub in ans:
                    tmp = Counter(sub)
                    for i in hmap:
                        if i not in tmp or tmp[i] < hmap[i]:
                            new = [element for element in sub] +[i]
                            next_ans.append(new)
                ans = next_ans
        return ans
```

```python
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        n = len(matrix)
        for i in range(n):
            for j in range(i+1,n):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
        i = 0
        j = n -1
        while i<j:
            for k in range(n):
                matrix[k][i], matrix[k][j] = matrix[k][j], matrix[k][i]
            i+=1
            j-=1
```

# 49. Group Anagrams

```python
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        hmap = {}
        for s in strs:
            sig = ''.join(sorted(list(s)))
            if sig not in hmap:
                hmap[sig] = [s]
            else:
                hmap[sig].append(s)
        return hmap.values()
        
```

# 50. Pow(x, n)

```python
class Solution(object):
    def myPow(self, x, n):
        """
        :type x: float
        :type n: int
        :rtype: float
        """
        if n == 0: return 1.0
        if n<0: 
            n = -n
            x = 1/x
        res = 1.0
        while n:
            if n%2!=0:
                res *= x
            x = x**2
            n = n//2
        return res
        
```
