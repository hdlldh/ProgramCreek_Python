# 51. N-Queens

```python
class Solution(object):
    def solveNQueens(self, n):
        """
        :type n: int
        :rtype: List[List[str]]
        """
        if n<=0: return []
        results= []
        self.dfs(n,[],results)
        print(results)
        ans = self.drawing(results,n)
        return ans
        
    def dfs(self,n,solution,results):
        if len(solution) ==n:
            results.append([s for s in solution])
            return
        
        for i in range(n):
            if self.isValid(solution,i):
                solution.append(i)
                self.dfs(n,solution,results)
                solution.pop()
                
                
    def isValid(self,solution, i):
        n = len(solution)
        if n==0: return True
        cur_row = n
        cur_col = i
        for row,col in enumerate(solution):
            if col == i: return False
            if row+col == cur_row+cur_col or row-col == cur_row-cur_col: return False
        return True
    
    def drawing(self,results,n):
        ans = []
        for result in results:
            tmp = []
            for num in result:
                s = ['.']*n
                s[num] = 'Q'
                tmp.append(''.join(s))
            ans.append(tmp)
        return ans
```

```python
class Solution(object):
    def totalNQueens(self, n):
        """
        :type n: int
        :rtype: int
        """
        
        if n<=0: return 0
        self.result = 0
        self.dfs(n,[])
        return self.result
        
    def dfs(self,n,solution):
        if len(solution) ==n:
            self.result+=1
            return
        
        for i in range(n):
            if self.isValid(solution,i):
                solution.append(i)
                self.dfs(n,solution)
                solution.pop()
                
                
    def isValid(self,solution, i):
        n = len(solution)
        if n==0: return True
        cur_row = n
        cur_col = i
        for row,col in enumerate(solution):
            if col == i: return False
            if row+col == cur_row+cur_col or row-col == cur_row-cur_col: return False
        return True
```

# 53. Maximum Subarray

```python
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        max_sofar = nums[0]
        max_sum = max_sofar
        for i in range(1,len(nums)):
            if max_sofar >0:
                max_sofar += nums[i]
            else:
                max_sofar = nums[i]
            max_sum = max(max_sum,max_sofar)
        
        return max_sum
            
```

# 56. Merge Intervals

```python
# Definition for an interval.
# class Interval(object):
#     def __init__(self, s=0, e=0):
#         self.start = s
#         self.end = e


class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[Interval]
        :rtype: List[Interval]
        """
        if not intervals: return []
        sorted_intervals = sorted(intervals,key=lambda x: x.start)
        ans = []
        anchor = 0
        for i in range(1,len(sorted_intervals)):
            if sorted_intervals[i].start <= sorted_intervals[anchor].end:
                sorted_intervals[anchor].end = max(sorted_intervals[anchor].end,sorted_intervals[i].end)
            else:
                ans.append(sorted_intervals[anchor])
                anchor = i
        ans.append(sorted_intervals[anchor])
        return ans
                
        
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def rotateRight(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        if not head: return head
        n = 0
        p = head
        while p:
            n+=1
            p = p.next
        k = (-k)%n
        slow = fast = head
        
        for i in range(k):
            fast = fast.next
            if not fast: fast = head
        if fast == slow: return head
        while slow.next != fast:
            slow = slow.next
        slow.next = None
        slow = fast
        while fast and fast.next:
            fast = fast.next
        fast.next = head
        return slow
        
        
            
```

# 66. Plus One

```python
class Solution(object):
    def plusOne(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """
        output = []
        carrier = 1
        for i in range(len(digits)-1,-1,-1):
            num = (digits[i] + carrier)
            d = num % 10
            output.insert(0,d)
            carrier = 1 if num >= 10 else 0
        if carrier: output.insert(0,1)
        return output
```

```python
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """
        m = len(a)
        n = len(b)
        a1 = list(a)
        a1.reverse()
        b1 = list(b)
        b1.reverse()
        c = []
        ci = '0'
        for i in range(max(m,n)):
            ai = a1[i] if i<m else '0'
            bi = b1[i] if i<n else '0'
            if ci=='0':
                if ai=='1' and bi=='1':
                    ci = '1'
                    c.append('0')
                elif ai=='1' or bi=='1':
                    c.append('1')
                else:
                    c.append('0')
            else:
                if ai=='0' and bi=='0':
                    ci = '0'
                    c.append('1')
                elif ai=='0' or bi=='0':
                    c.append('0')
                else:
                    c.append('1')
        if ci=='1':c.append('1')
        c.reverse()
        return ''.join(c)
                
```

# 70. Climbing Stairs

```python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n ==1: return 1
        if n ==2: return 2
        
        memo = [0]*(n+1)
        memo[1] =1
        memo[2] =2
        for i in range(3,n+1):
            memo[i] = memo[i-1] + memo[i-2]
        return memo[n]
        
```

```python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        
        memo = [0]*(n+1)
        
        return self.helper(memo,n)
        
    
    def helper(self,memo,n):
        if n==1: return 1
        if n==2: return 2
        if memo[n] >0: return memo[n]
        memo[n] = self.helper(memo,n-1)+self.helper(memo,n-2)
        return memo[n]
            
```

# 74. Search a 2D Matrix

```python
class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        if not matrix: return False
        m, n = len(matrix), len(matrix[0])
        low = 0
        high = m*n-1
        while low<=high:
            mid = (low+high)//2
            i,j = divmod(mid,n)
            if matrix[i][j]==target:
                return True
            elif matrix[i][j]<target:
                low = mid+1
            else:
                high = mid-1
        return False
```

# 89. Gray Code

```python
class Solution(object):
    def grayCode(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        if n ==0: return [0]
        ans = [0,1]
        if n==1: return ans
        for i in range(n-1):
            next_ans = []
            for a in ans:
                next_ans.append(a*2)
            for a in reversed(ans):
                next_ans.append(a*2+1)
            ans = next_ans
            
        return ans
```

```python
class Solution(object):
    def restoreIpAddresses(self, s):
        """
        :type s: str
        :rtype: List[str]
        """
        ans = []
        self.helper(s,0,"",ans)
        return ans
        
        
    def helper(self, s, n, out, ans):
        if n == 4:
            if not s: ans.append(out)
            return
        for k in range(1,4):
            if len(s) < k: continue
            tmp = int(s[0:k])
            if tmp>255 or len(str(tmp))!=k:continue
            if n==3:
                self.helper(s[k:],n+1,out+s[0:k],ans)
            else:
                self.helper(s[k:],n+1,out+s[0:k]+'.',ans)
```

# 94. Binary Tree Inorder Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        ans = []
        stack = []
        if not root: return ans
        cur = root
        while cur or stack:
            while cur:
                stack.append(cur)
                cur = cur.left
            cur = stack.pop()
            ans.append(cur.val)
            cur = cur.right
        return ans
            
            
            
```

# 95. Unique Binary Search Trees II

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution(object):
    def generateTrees(self, n):
        """
        :type n: int
        :rtype: List[TreeNode]
        """
        ans = self.helper(1,n)
        return ans
            
    def helper(self,l,r):
        if l>r: return []
        result = []
        if l==r:
            result.append(TreeNode(l))
            return result
        for i in range(l,r+1):
            for left in self.helper(l,i-1) or [None]:
                for right in self.helper(i+1,r) or [None]:
                    node = TreeNode(i)
                    node.left = left
                    node.right = right
                    result.append(node)
        return result
                
```

# 96. Unique Binary Search Trees

```python
class Solution(object):
    def numTrees(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n == 0: return 1
        if n == 1: return 1
        if n == 2: return 2
        dp = [0]*(n+1)
        dp[0] = 1
        dp[1] = 1
        dp[2] = 2
        for i in range(3,n+1):
            for j in range(1,i+1):
                dp[i] += dp[j-1]*dp[i-j]
        return dp[n]
            
```

# 98. Validate Binary Search Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution(object):
    def isValidBST(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        self.prev = None
        self.valid = True
        self.helper(root)
        return self.valid
        
    def helper(self,root):
        if not root: return
        self.helper(root.left)
        if self.prev is not None and root.val<=self.prev:
            self.valid = False
        self.prev = root.val
        self.helper(root.right)
```

# 100. Same Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isSameTree(self, p, q):
        """
        :type p: TreeNode
        :type q: TreeNode
        :rtype: bool
        """
        
        if p and not q or q and not p: return False
        if not p and not q: return True
        if p.val != q.val: return False
        return self.isSameTree(p.left,q.left) and self.isSameTree(p.right,q.right)
    
   
```
