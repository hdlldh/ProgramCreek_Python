# 653. Two Sum IV - Input is a BST

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def findTarget(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: bool
        """
        self.hashset = set()
        return self.helper(root,k)
    
    def helper(self,root,k):
        if not root: return False
        if root.val in self.hashset: return True
        self.hashset.add(k-root.val)
        return self.helper(root.left,k) or self.helper(root.right,k)
        
```

# 657. Robot Return to Origin

```python
class Solution(object):
    def judgeCircle(self, moves):
        """
        :type moves: str
        :rtype: bool
        """
        counter = {'U':0,'D':0,'L':0,'R':0}
        for m in moves:
            counter[m] += 1
        if counter['U']==counter['D'] and counter['L']==counter['R']:
            return True
        return False
        
```

# 661. Image Smoother

```python
class Solution(object):
    def imageSmoother(self, M):
        """
        :type M: List[List[int]]
        :rtype: List[List[int]]
        """
        r,c = len(M), len(M[0])
        ans = [[0]*c for _ in range(r)]
        for i in range(r):
            for j in range(c):
                cnt = 0
                for k in range(max(0,i-1),min(r,i+2)):
                    for l in range(max(0,j-1),min(c,j+2)):
                        cnt += 1
                        ans[i][j] += M[k][l]
                ans[i][j] = ans[i][j]//cnt
        return ans
```

# 669. Trim a Binary Search Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def trimBST(self, root, L, R):
        """
        :type root: TreeNode
        :type L: int
        :type R: int
        :rtype: TreeNode
        """
        return self.trim(root,L,R)
        
    def trim(self,root,L,R):
        if not root: return None
        if root.val > R:
            return self.trim(root.left,L,R)
        if root.val < L:
            return self.trim(root.right,L,R)
        root.left = self.trim(root.left,L,R)
        root.right = self.trim(root.right,L,R)
        return root
```

# 674. Longest Continuous Increasing Subsequence

```python
class Solution(object):
    def findLengthOfLCIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        ans = 0
        anchor = 0
        for i in range(len(nums)):
            if i and nums[i]<nums[i-1]:
                anchor = i
            ans = max(ans,i-anchor+1)
        return ans
                
```

```python
class Solution(object):
    def findLengthOfLCIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        if n==0:return 0
        dp = [1]*n
        for i in range(1,n):
            if i>0 and nums[i]>nums[i-1]:
                dp[i] = dp[i-1] +1
        return max(dp)
```

# 680. Valid Palindrome II

```python
class Solution(object):
    def validPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        n = len(s)
        i = 0
        j = n-1
        while i<j:
            if s[i]!=s[j]:
                return self.isPali(s[i+1:j+1]) or self.isPali(s[i:j])
            i += 1
            j -= 1
        return True
        
    def isPali(self,s):
        if not s: return True
        i = 0
        j = len(s)-1
        while i<j:
            if s[i]!=s[j]: return False
            i+=1
            j-=1
        return True
```

# 682. Baseball Game

```python
class Solution(object):
    def calPoints(self, ops):
        """
        :type ops: List[str]
        :rtype: int
        """
        stack = []
        for op in ops:
            if op == '+':
                stack.append(stack[-1]+stack[-2])
            elif op == 'D':
                stack.append(stack[-1]*2)
            elif op == 'C':
                stack.pop()
            else:
                stack.append(int(op))
        return sum(stack)
```

# 687. Longest Univalue Path

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def longestUnivaluePath(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.longest = 0
        self.arrow_length(root)
        return self.longest
        
        
    def arrow_length(self, root):
        if not root: return 0
        left_cnt = self.arrow_length(root.left)
        right_cnt = self.arrow_length(root.right)
        
        if root.left and root.left.val == root.val: left_cnt +=1
        else: left_cnt = 0
        if root.right and root.right.val == root.val: right_cnt +=1
        else: right_cnt =0
        self.longest = max(self.longest,left_cnt+right_cnt)
        
        return max(left_cnt,right_cnt)
        
```

# 690. Employee Importance

```python
"""
# Employee info
class Employee(object):
    def __init__(self, id, importance, subordinates):
        # It's the unique id of each node.
        # unique id of this employee
        self.id = id
        # the importance value of this employee
        self.importance = importance
        # the id of direct subordinates
        self.subordinates = subordinates
"""
class Solution(object):
    def getImportance(self, employees, id):
        """
        :type employees: Employee
        :type id: int
        :rtype: int
        """
        self.emap = {e.id: e for e in employees}
        return self.dfs(id)
    
    def dfs(self,id):
        e = self.emap[id]
        if not e.subordinates: return e.importance
        return e.importance + sum([self.dfs(i) for i in e.subordinates])
        
```

# 693. Binary Number with Alternating Bits

```python
class Solution(object):
    def hasAlternatingBits(self, n):
        """
        :type n: int
        :rtype: bool
        """
        lastbit = n & 1
        while n:
            n = n >>1
            currbit = n&1
            if lastbit == currbit:return False
            lastbit = currbit
        return True
            
```

# 696. Count Binary Substrings

```python
class Solution(object):
    def countBinarySubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        n = len(s)
        cnt = 0
        ones = 0
        zeros = 0
        for i in range(n):
            if i == 0:
                if s[0] =='0': 
                    zeros+=1
                else: 
                    ones+=1
            else:
                if s[i] == '0' :
                    if s[i-1]=='0':
                        zeros += 1
                    else:
                        zeros = 1
                    if ones>=zeros: cnt += 1
                else:
                    if s[i-1] == '1':
                        ones += 1
                    else:
                        ones = 1
                    if zeros>=ones: cnt += 1
        return cnt       
```

# 697. Degree of an Array

```python
class Solution(object):
    def findShortestSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        hashmap = {}
        n = len(nums)
        ans = n
        degree = 1
        for i in range(n):
            if nums[i] not in hashmap:
                hashmap[nums[i]] = [i,i,1]
            else:
                hashmap[nums[i]][1] = i
                hashmap[nums[i]][2] += 1
                degree = max(degree,hashmap[nums[i]][2])
        for k, v in hashmap.items():
            if v[2] == degree:
                ans = min(ans(v[1]-v[0]+1))
        return ans
            
        
        
```

# 700. Search in a Binary Search Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def searchBST(self, root, val):
        """
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode
        """
        if not root: return None
        if root.val== val: return root
        if root.val< val: return self.searchBST(root.right,val)
        if root.val> val: return self.searchBST(root.left,val)
```
