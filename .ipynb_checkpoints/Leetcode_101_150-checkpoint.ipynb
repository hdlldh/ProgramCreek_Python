# 101. Symmetric Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if root is None: return True
        else:
            return self.isMirror(root.left,root.right)
                
    def isMirror(self, left, right):
        if left is None and right is None:
            return True
        if left is None or right is None:
            return False
        return left.val == right.val and self.isMirror(left.left,right.right) and self.isMirror(left.right,right.left)
                    
            
```

```python
from collections import deque
class Solution:
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        q = deque()
        q.append(root)
        q.append(root)
        while len(q)>0:
            t1 = q.popleft()
            t2 = q.popleft()
            if (t1 is None and t2 is None): continue
            if (t1 is None or t2 is None): return False
            if (t1.val != t2.val): return False
            q.append(t1.left)
            q.append(t2.right)
            q.append(t1.right)
            q.append(t2.left)
        return True
        
            
```

# 102. Binary Tree Level Order Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        ans = []
        level = [root]
        while level:
            next_level = []
            cur_vals = []
            for node in level:
                cur_vals.append(node.val)
                if node.left: next_level.append(node.left)
                if node.right: next_level.append(node.right)
            ans.append(cur_vals)
            level = next_level
        return ans
```

# 103. Binary Tree Zigzag Level Order Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def zigzagLevelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        ans = []
        level = [root]
        flip = False
        while level:
            next_level = []
            cur_vals = []
            for node in level:
                cur_vals.append(node.val)
                if node.left: next_level.append(node.left)
                if node.right: next_level.append(node.right)
            if flip: cur_vals.reverse()
            ans.append(cur_vals)
            flip = not flip
            level = next_level
        return ans
```

# 104. Maximum Depth of Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
        

class Solution:
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None: return 0
        return 1+ max(self.maxDepth(root.left),self.maxDepth(root.right))
```

# 105. Construct Binary Tree from Preorder and Inorder Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: TreeNode
        """
        if inorder:
            val = preorder.pop(0)
            root = TreeNode(val)
            idx = inorder.index(val)
            root.left = self.buildTree(preorder,inorder[:idx])
            root.right = self.buildTree(preorder,inorder[idx+1:])
            return root
        return None
```

# 106. Construct Binary Tree from Inorder and Postorder Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def buildTree(self, inorder, postorder):
        """
        :type inorder: List[int]
        :type postorder: List[int]
        :rtype: TreeNode
        """
        if inorder:
            val = postorder.pop()
            root = TreeNode(val)
            idx = inorder.index(val)
            root.right = self.buildTree(inorder[idx+1:],postorder)
            root.left = self.buildTree(inorder[:idx],postorder)
            return root
        return None
```

# 107. Binary Tree Level Order Traversal II

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def levelOrderBottom(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        output = []
        level = [root]
        value = [root.val]
        while level:
            output.insert(0,value)
            next_level = []
            next_value = []
            for node in level:
                if node.left: 
                    next_level.append(node.left)
                    next_value.append(node.left.val)
                if node.right: 
                    next_level.append(node.right)
                    next_value.append(node.right.val)
            level = next_level
            value = next_value
        return output
            
            
```

# 108. Convert Sorted Array to Binary Search Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def sortedArrayToBST(self, nums):
        """
        :type nums: List[int]
        :rtype: TreeNode
        """
        return self.buildTree(nums)
        
    def buildTree(self,nums):
        len_nums = len(nums)
        if len_nums == 0:
            return None
        elif len_nums == 1:
            return TreeNode(nums[0])
        else:
            mid = len_nums // 2
            node = TreeNode(nums[mid])
            node.left = self.buildTree(nums[:mid])
            node.right = self.buildTree(nums[mid+1:])
            return node
```

# 109. Convert Sorted List to Binary Search Tree

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def sortedListToBST(self, head):
        """
        :type head: ListNode
        :rtype: TreeNode
        """
        if not head: return None
        mid = self.find_middle(head)
        cur = TreeNode(mid.val)
        if head == mid: return cur
        cur.left = self.sortedListToBST(head)
        cur.right = self.sortedListToBST(mid.next)
        return cur
        
    def find_middle(self,head):
        prev = None
        slow = head
        fast = head
        while fast and fast.next:
            prev = slow
            slow = slow.next
            fast = fast.next.next
        if prev: prev.next = None
        return slow
```

# 110. Balanced Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isBalanced(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        return self.maxHeight(root)!=-1
        
    def maxHeight(self, node):
        if node is None: return 0
        leftHeight = self.maxHeight(node.left)
        rightHeight = self.maxHeight(node.right)
        if leftHeight == -1 or rightHeight== -1: return -1
        if abs (leftHeight - rightHeight) >1: 
            return -1
        else:
            return 1+ max(leftHeight,rightHeight)
```

# 111. Minimum Depth of Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


# BFS method

from collections import deque
class Solution:
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root: return 0
        node_queue = deque([(root,1),])
        while node_queue:
            node, depth = node_queue.popleft()
            children = [node.left, node.right]
            if not any(children):
                return depth
            for c in children:
                if c: node_queue.append((c,depth+1))
```

```python
###  recursive method
class Solution:
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root: return 0
        children = [root.left,root.right]
        if not any(children): return 1
        depth = float('inf')
        for c in children:
            if c: depth = min(self.minDepth(c),depth)
        return depth + 1
        
```

```python
# DFS method

class Solution:
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root: return 0
        node_stack = [(root,1),]
        min_depth = float('inf')
        while node_stack:
            node, depth = node_stack.pop()
            children = [node.left, node.right]
            if not any(children):
                min_depth = min(min_depth,depth)
            for c in children:
                if c: node_stack.append((c,depth+1))
        return min_depth
```

# 112. Path Sum

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def hasPathSum(self, root, sum):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: bool
        """
        if not root: return False
        sum = sum - root.val
        if not root.left and not root.right: return  sum ==0
        return self.hasPathSum(root.left,sum) or self.hasPathSum(root.right,sum) 
         
```

```python
class Solution:
    def hasPathSum(self, root, sum):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: bool
        """
        if not root: return False
        stack = [(root,sum-root.val)]
        while stack:
            node, bal = stack.pop()
            if node.left is None and node.right is None and bal==0: return True
            if node.left: stack.append((node.left,bal-node.left.val))
            if node.right: stack.append((node.right,bal-node.right.val))
        return False
```

# 113. Path Sum II

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def pathSum(self, root, sum):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: List[List[int]]
        """
        if not root: return []
        output = []
        stack = [(root,sum-root.val,[root.val])]
        while stack:
            node, bal, value_list = stack.pop()
            if not node.left and not node.right and bal==0:
                output.append(value_list)
            else:
                if node.left:
                    value_list_copy = [item for item in value_list]
                    value_list_copy.append(node.left.val)
                    stack.append((node.left,bal-node.left.val,value_list_copy))
                if node.right:
                    value_list_copy = [item for item in value_list]
                    value_list_copy.append(node.right.val)
                    stack.append((node.right,bal-node.right.val,value_list_copy))
        return output
```

# 116. Populating Next Right Pointers in Each Node

```python
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, left, right, next):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
"""
class Solution(object):
    def connect(self, root):
        """
        :type root: Node
        :rtype: Node
        """
        if not root: return root
        level = [root]
        while level:
            next_level = []
            node_next = None
            for i in range(len(level)):
                if i!=len(level)-1:
                    level[i].next = level[i+1]
                if level[i].left:
                    next_level.append(level[i].left)
                    next_level.append(level[i].right)
            level = next_level
        return root
```

# 120. Triangle

```python
class Solution(object):
    def minimumTotal(self, triangle):
        """
        :type triangle: List[List[int]]
        :rtype: int
        """
        n = len(triangle)
        for i in range(1,n):
            for j in range(i+1):
                if j==0:
                    triangle[i][j] += triangle[i-1][j]
                elif j == i:
                    triangle[i][j] += triangle[i-1][j-1]
                else:
                    triangle[i][j] += min(triangle[i-1][j],triangle[i-1][j-1]) 
        return min(triangle[n-1])
        
        
```

```python
class Solution(object):
    def minimumTotal(self, triangle):
        """
        :type triangle: List[List[int]]
        :rtype: int
        """
        n = len(triangle)
        res = [0]*n
        res[0] = triangle[0][0]
        for i in range(1,n):
            next_res = [0]*n
            for j in range(i+1):
                if j==0:
                    next_res[j] = triangle[i][j] + res[j]
                elif j == i:
                    next_res[j] = triangle[i][j] + res[j-1]
                else:
                    next_res[j] = triangle[i][j]+min(res[j],res[j-1])
            res= next_res
        return min(res)
```

# 121. Best Time to Buy and Sell Stock

```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        n = len(prices)
        if n == 0: return 0
        f = [0]*n
        minV = float('inf')
        for i in range(n):
            if i>0:
                f[i] = prices[i]-minV
            minV=min(minV,prices[i])
        return max(f)
```

# 123. Best Time to Buy and Sell Stock III

```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        n = len(prices)
        if n==0: return 0
        f = [[0]*5 for _ in range(n+1)]
        for i in range(1,5):
            f[0][i] = -float('inf')
        
        for i in range(1,n+1):
            for j in range(5):
                ## doesn't own any stock or sell on i-th day
                if j%2==0:
                    if i>1 and j>=1:
                        f[i][j] = max(f[i-1][j],f[i-1][j-1]+prices[i-1]-prices[i-2])
                ## own stock or buy on i-th day
                else:
                    if i>1 and j>=1:
                        f[i][j] = max(f[i-1][j]+prices[i-1]-prices[i-2],f[i][j-1])
                        
        ans = max(f[n][0],f[n][2],f[n][4])
        return ans
```

# 125. Valid Palindrome

```python
class Solution:
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        slist = [c.lower() for c in s if c.isalnum()]
        i = 0
        j = len(slist)-1
        while (i<=j):
            if slist[i]!=slist[j]: return False
            i += 1
            j -= 1
        return True
   
```

# 127. Word Ladder

```python
class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """
        if not beginWord or not endWord or not wordList: return 0
        if endWord not in wordList: return 0
        
        L = len(beginWord)
        wordList = set(wordList)
        tovisit = [beginWord]
        distance = {}
        distance[beginWord] = 1
        while tovisit:
            word = tovisit.pop(0)
            for i in range(L):
                for ch in range(ord('a'),ord('z')+1):
                    if chr(ch) == word[i]:continue
                    newWord = word[:i]+chr(ch)+word[i+1:]
                    if newWord == endWord: return distance[word]+1
                    if newWord in wordList and newWord not in distance:
                        distance[newWord] = distance[word]+1
                        tovisit.append(newWord)          
        
        return 0
```

# 129. Sum Root to Leaf Numbers

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def sumNumbers(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.ans = 0
        self.helper(root,0)
        return self.ans
        
    def helper(self,root, inc):
        if not root: return
        if not root.left and not root.right:
            self.ans += inc*10+root.val
        else:
            if root.left:
                self.helper(root.left,inc*10+root.val)
            if root.right:
                self.helper(root.right,inc*10+root.val)
```

# 131. Palindrome Partitioning

```python
class Solution(object):
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        n = len(s)
        if n==0: return []
        ans = []
        memo = [[False]*n for _ in range(n)]
        for c in range(n):
            i=j=c
            while i>=0 and j<n:
                if s[i]==s[j]:
                    memo[i][j] = True
                    i-=1
                    j+=1
                else:
                    break
        for c in range(1,n):
            i = c-1
            j = c
            while i>=0 and j<n:
                if s[i]==s[j]:
                    memo[i][j] = True
                    i-=1
                    j+=1
                else:
                    break
        self.helper(s,0,memo,[],ans)
        return ans
                    
        
        
    def helper(self,s,i,memo,out,ans):
        n = len(s)
        if i>=n:
            ans.append(out)
            return
        for j in range(i,n):
            if memo[i][j]:
                self.helper(s,j+1,memo,out+[s[i:j+1]],ans)
```

# 132. Palindrome Partitioning II

```python
class Solution(object):
    def minCut(self, s):
        """
        :type s: str
        :rtype: int
        """
        n = len(s)
        if n==0: return 0
        memo = [[False]*n for _ in range(n)]
        for c in range(n):
            i=j=c
            while i>=0 and j<n:
                if s[i]==s[j]:
                    memo[i][j] = True
                    i-=1
                    j+=1
                else:
                    break
        for c in range(1,n):
            i = c-1
            j = c
            while i>=0 and j<n:
                if s[i]==s[j]:
                    memo[i][j] = True
                    i-=1
                    j+=1
                else:
                    break
        dp = [0]*(n+1)
        for i in range(1,n+1):
            dp[i] = float('inf')
            for j in range(i):
                if memo[j][i-1]:
                    dp[i] = min(dp[i],dp[j]+1)
        return dp[-1]-1
```

# 133. Clone Graph

```python
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, neighbors):
        self.val = val
        self.neighbors = neighbors
"""
class Solution(object):
    def cloneGraph(self, node):
        """
        :type node: Node
        :rtype: Node
        """
        if not node: return node
        visited = set()
        created = {}
        tovisit = [node]
        while tovisit:
            nd = tovisit.pop()
            visited.add(nd)
            nb_copy =[]
            for nb in nd.neighbors:
                if nb not in created:
                    new = Node(None,None)
                    created[nb] = new
                nb_copy.append(created[nb])
                if nb not in visited:
                    tovisit.append(nb)
                
            if nd not in created:
                new = Node(None,None)
                created[nd] = new
            created[nd].val = nd.val
            created[nd].neighbors = nb_copy
        return created[node]
```

# 134. Gas Station

```python
class Solution(object):
    def canCompleteCircuit(self, gas, cost):
        """
        :type gas: List[int]
        :type cost: List[int]
        :rtype: int
        """
        if not gas: return -1
        n = len(gas)
        total_tank, cur_tank = 0, 0
        starting_station = 0 
        for i in range(n):
            total_tank += gas[i] - cost[i]
            cur_tank += gas[i] - cost[i]
            if cur_tank<0:
                starting_station = i+1
                cur_tank = 0
                
        return starting_station if total_tank>=0 else -1
        
```

# 135. Candy

```python
class Solution(object):
    def candy(self, ratings):
        """
        :type ratings: List[int]
        :rtype: int
        """
        if not ratings: return 0
        
        n = len(ratings)
        left2right = [1]*n
        right2left = [1]*n
        for i in range(1, n):
            if ratings[i]>ratings[i-1]:
                left2right[i] = left2right[i-1]+1
        
            
        for i in range(n-1,0,-1):
            if ratings[i-1]>ratings[i]:
                right2left[i-1] = right2left[i]+1
                
        for i in range(n):
            left2right[i] = max(left2right[i],right2left[i])
            
        return sum(left2right)
```

# 139. Word Break

```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """
        wordSet = set(wordDict)
        n = len(s)
        dp = [False]*(n+1)  # dp[i]: s[:i] can be segmented
        dp[0] = True
        for i in range(1,n+1):
            for j in range(i):
                if dp[j] and s[j:i] in wordSet:
                    dp[i] = True
        return dp[-1]    
```

# 140. Word Break II

```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: List[str]
        """
        wordSet = set(wordDict)
        ans = []
        self.helper(s,"",wordSet,ans)
        return ans
        
    
    def helper(self,s, out, wordSet,ans):
        if s in wordSet:
            ans.append(out+s)
            return
        n = len(s)
        for i in range(1,n):
            if s[:i] in wordSet:
                self.helper(s[i:],out+' '+s[:i],wordSet,ans)
        
```

# 141. Linked List Cycle

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        if not head: return False
        hashset = set()
        
        while head:
            if head not in hashset:
                hashset.add(head)
                head = head.next
            else:
                return True
        return False
                
                
        
```

# 144. Binary Tree Preorder Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        pre = []
        stack = [root]
        while stack:
            node = stack.pop()
            pre.append(node.val)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
        return pre
```

# 145. Binary Tree Postorder Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def postorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        post = []
        stack = [root]
        while stack:
            node = stack.pop()
            post.insert(0,node.val)
            if node.left: stack.append(node.left)
            if node.right: stack.append(node.right)
        return post
```

# 146. LRU Cache

```python
class LRUCache(object):
    class Node():
        def __init__(self,key,value,prev=None,next=None):
            self.key = key
            self.value = value
            self.prev = prev
            self.next = next
        

    def __init__(self, capacity):
        """
        :type capacity: int
        """
        self.capacity = capacity
        self.head = self.Node(None,None)
        self.tail = self.Node(None,None)
        self.head.next = self.tail
        self.tail.prev = self.head
        self.map = {}
        
    def add_last(self,p):
        last = self.tail.prev
        last.next = p
        p.prev = last
        self.tail.prev = p
        p.next = self.tail
        
    def remove_first(self):
        first = self.head.next
        self.head.next = first.next
        first.next.prev = self.head
        return first
        

    def get(self, key):
        """
        :type key: int
        :rtype: int
        """
        if key in self.map:
            p = self.map[key]
            p.prev.next = p.next
            p.next.prev = p.prev
            self.add_last(p)
            return p.value
        else:
            return -1
       

    def put(self, key, value):
        """
        :type key: int
        :type value: int
        :rtype: None
        """
        if key not in self.map:
            if len(self.map)==self.capacity:
                first = self.remove_first()
                del self.map[first.key]
                
            p = self.Node(key, value)
            self.map[key] = p
            self.add_last(p)
        else:
            p = self.map[key]
            p.value = value
            p.prev.next = p.next
            p.next.prev = p.prev
            self.add_last(p)
```

# 147. Insertion Sort List

```python
class Solution(object):
    def insertionSortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head: return head
        dummy = ListNode(None)
        dummy.next = head
        
        while head and head.next:
            if head.val > head.next.val:
                toInsert = head.next
                head.next = head.next.next
                p = dummy
                while p.next.val < toInsert.val:
                    p = p.next
                toInsert.next = p.next
                p.next = toInsert
            else:
                head = head.next
            
        return dummy.next
```

# 149. Max Points on a Line

```python
# Definition for a point.
# class Point(object):
#     def __init__(self, a=0, b=0):
#         self.x = a
#         self.y = b

class Solution(object):
    def maxPoints(self, points):
        """
        :type points: List[Point]
        :rtype: int
        """
        n = len(points)
        if n==0:return 0
        def gcd(a,b):
            return a if b==0 else gcd(b,a%b)
        
        res = 0
        for i in range(n):
            duplicates = 1
            hashmap = {} ## record # of far-end points with certain (dx, dy)
            for j in range(i+1,n):
                if i==j: continue
                if points[i].x==points[j].x and points[i].y==points[j].y:
                    duplicates += 1
                    continue
                else:
                    dx = points[j].x - points[i].x
                    dy = points[j].y - points[i].y
                    if dx <0:
                        dx = -dx
                        dy = -dy
                    d = gcd(dx,dy)
                    
                    dx = dx//d
                    dy = dy//d
                    
                    if (dx,dy) not in hashmap:
                        hashmap[(dx,dy)] = 1
                    else:
                        hashmap[(dx,dy)] +=1
            res = max(res,duplicates)
            for v in hashmap.values():
                res=max(res,duplicates+v)
            
        return res
```

```python
class Point(object):
    def __init__(self, a=0, b=0):
        self.x = a
        self.y = b
points= [[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]
pp = []
for p in points:
    pp.append(Point(p[0],p[1]))
a = Solution()
a.maxPoints(pp)
```

# 150. Evaluate Reverse Polish Notation

```python
class Solution(object):
    def evalRPN(self, tokens):
        """
        :type tokens: List[str]
        :rtype: int
        """
        stack = []
        for t in tokens:
            if t not in ['+','-','*','/']:
                stack.append(int(t))
            else:
                p1 = stack.pop()
                p2 = stack.pop()
                if t=='+':
                    p3 = p1+p2
                elif t=='-':
                    p3 = p2-p1
                elif t== '*':
                    p3 = p1*p2
                else:
                    if p1*p2 < 0 and p2 % p1 != 0:
                        p3 = p2//p1+1
                    else:
                        p3 = p2//p1
                stack.append(p3)
                
        return stack[0]
```
