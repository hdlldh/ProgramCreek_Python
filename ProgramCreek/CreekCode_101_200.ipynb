# 313. Super Ugly Number

```python
class Solution(object):
    def nthSuperUglyNumber(self, n, primes):
        """
        :type n: int
        :type primes: List[int]
        :rtype: int
        """
        
        heap = [1]
        hset = set()
        hset.add(1)
        for i in range(n):
            num = heapq.heappop(heap)
            if i==n-1: return num
            for p in primes:
                if p*num not in hset:
                    hset.add(p*num)
                    heapq.heappush(heap,p*num)
        
```

# 373. Find K Pairs with Smallest Sums

```python
class Solution(object):
    def kSmallestPairs(self, nums1, nums2, k):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :type k: int
        :rtype: List[List[int]]
        """
        n1 = len(nums1)
        n2 = len(nums2)
        heap = []
        for i in range(n1):
            for j in range(n2):
                heapq.heappush(heap,(-nums1[i]-nums2[j],i,j))
                if len(heap)> k:
                    heapq.heappop(heap)
        ans = []
        while heap:
            _,i,j = heapq.heappop(heap)
            ans.append([nums1[i],nums2[j]])
        ans.reverse()
        return ans
        
```

# 189. Rotate Array

```python
class Solution(object):
    def rotate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: void Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        k = k%n
        
        def reverse(nums,start,end):
            
            while start<end:
                nums[start],nums[end] = nums[end],nums[start]
                start += 1
                end -= 1
        
        reverse(nums,0,n-1)
        reverse(nums,0,k-1)
        reverse(nums,k,n-1)
```

# 74. Search a 2D Matrix

```python
class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        if not matrix: return False
        m, n = len(matrix), len(matrix[0])
        low = 0
        high = m*n-1
        while low<=high:
            mid = (low+high)//2
            i,j = divmod(mid,n)
            if matrix[i][j]==target:
                return True
            elif matrix[i][j]<target:
                low = mid+1
            else:
                high = mid-1
        return False
```

# 378. Kth Smallest Element in a Sorted Matrix

```python
class Solution(object):
    def kthSmallest(self, matrix, k):
        """
        :type matrix: List[List[int]]
        :type k: int
        :rtype: int
        """
        n = len(matrix)
        
        def count(matrix, target):
            n = len(matrix)
            i = n-1
            j = 0
            cnt = 0
            while i>=0 and j<n:
                if matrix[i][j]<=target:
                    cnt += i+1
                    j += 1
                else:
                    i-=1
            return cnt
        low = matrix[0][0]
        high = matrix[n-1][n-1]
        
        while low<=high:
            mid = (low+high)//2
            cnt = count(matrix,mid)
            if cnt>=k:
                high = mid -1
            else:
                low = mid +1
        return low
```

# 353. Design Snake Game

```python
class SnakeGame(object):

    def __init__(self, width, height, food):
        """
        Initialize your data structure here.
        @param width - screen width
        @param height - screen height 
        @param food - A list of food positions
        E.g food = [[1,1], [1,0]] means the first food is positioned at [1,1], the second is at [1,0].
        :type width: int
        :type height: int
        :type food: List[List[int]]
        """
        self.width = width
        self.height = height
        self.queue = [[0,0]]
        self.food = food
        self.score = 0
        

    def move(self, direction):
        """
        Moves the snake.
        @param direction - 'U' = Up, 'L' = Left, 'R' = Right, 'D' = Down 
        @return The game's score after the move. Return -1 if game over. 
        Game over when snake crosses the screen boundary or bites its body.
        :type direction: str
        :rtype: int
        """
        x, y = self.queue[0]
        if direction == 'U': 
            x -= 1
            if x == -1: return -1
        if direction == 'L': 
            y -= 1
            if y == -1: return -1
        if direction == 'R': 
            y += 1
            if y == self.width: return -1
        if direction == 'D': 
            x += 1
            if x == self.height: return -1
        
        if [x,y] in self.queue[:-1]: return -1
        
        self.queue.insert(0,[x,y])
        
        if self.food and [x,y] == self.food[0]:
            self.score += 1
            self.food.pop(0)
        else:
            self.queue.pop()
        return self.score
```

# 305. Number of Islands II

```python
class Solution(object):
    def numIslands2(self, m, n, positions):
        """
        :type m: int
        :type n: int
        :type positions: List[List[int]]
        :rtype: List[int]
        """
        
        parent = [-1]*(m*n)
        size = [0]*(m*n)
        filled = set()
        ans = []
        directions = [[-1,0],[1,0],[0,-1],[0,1]]
        count = 0
        
        def add(parent,size,i):
            parent[i] = i
            size[i] += 1
        
        def find(parent,i):
            if parent[i]!=i:
                return find(parent,parent[i])
            return i
        
        def union(parent,size,i,j):
            root_i = find(parent,i)
            root_j = find(parent,j)
            
            if size[root_i]>=size[root_j]:
                parent[root_j] = root_i
                size[root_i] += size[root_j]
            else:
                parent[root_i] = root_j
                size[root_j] += size[root_i]
            return
        
        for pos in positions:
            index = pos[0]*n+pos[1]
            add(parent,size,index)
            if (pos[0],pos[1]) in filled: 
                ans.append(count)
                continue
            filled.add((pos[0],pos[1]))
            count += 1
            
            for d in directions:
                i = pos[0]+d[0]
                j = pos[1]+d[1]
                if i<0 or i>=m: continue
                if j<0 or j>=n: continue
                if (i,j) not in filled: continue
                if find(parent,i*n+j)!=find(parent,index):
                    union(parent,size,i*n+j,index)
                    count -= 1
            ans.append(count)
        return ans
```

# 323. Number of Connected Components in an Undirected Graph

```python
class Solution(object):
    def countComponents(self, n, edges):
        """
        :type n: int
        :type edges: List[List[int]]
        :rtype: int
        """
        parent = [i for i in range(n)]
        rank = [1]* n
        
        def find(parent,i):
            if parent[i]!=i:
                return find(parent,parent[i])
            return i
        
        def union(parent,size,i,j):
            root_i = find(parent,i)
            root_j = find(parent,j)
            if root_i != root_j:
                if size[root_i]>=size[root_j]:
                    parent[root_j] = root_i
                    size[root_i] += size[root_j]
                else:
                    parent[root_i] = root_j
                    size[root_j] += size[root_i]
            return
        
        for edge in edges:
            i, j = edge
            union(parent,rank,i,j)
        
        ans = 0
        for i in range(n):
            if parent[i] == i:
                ans+= 1
        return ans
```

# 947. Most Stones Removed with Same Row or Column

```python
class Solution(object):
    def removeStones(self, stones):
        """
        :type stones: List[List[int]]
        :rtype: int
        """
        n = len(stones)
        if n==0:return 0
        
        
        parent = [i for i in range(20000)]
        rank = [1]* 20000
        
        def find(parent,i):
            if parent[i]!=i:
                return find(parent,parent[i])
            return i
        
        def union(parent,rank,i,j):
            root_i = find(parent,i)
            root_j = find(parent,j)
            if root_i==root_j: return
            if rank[root_i]>=rank[root_j]:
                parent[root_j] = root_i
                rank[root_i] += rank[root_j]
            else:
                parent[root_i] = root_j
                rank[root_j] += rank[root_i]
            return
        
        for s in stones:
            union(parent,rank,s[0],s[1]+10000)
        
        roots = set()
        for s in stones:
            roots.add(find(parent,s[0]))
        return n - len(roots)
```

# 329. Longest Increasing Path in a Matrix

```python
class Solution(object):
    def longestIncreasingPath(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        m = len(matrix)
        if m==0: return 0
        n = len(matrix[0])
        mem = [[0]*n for _ in range(m)]
        ans = 1
        
        directions = [[-1,0],[1,0],[0,-1],[0,1]]
        
        def dfs(matrix,i,j,mem):
            if mem[i][j]>0: return mem[i][j]
            m,n = len(matrix),len(matrix[0])
            
            for d in directions:
                k = i+d[0]
                l = j+d[1]
                if k<0 or k>=m or l<0 or l>=n: continue
                if matrix[i][j]>matrix[k][l]:
                    mem[i][j] = max(mem[i][j],dfs(matrix,k,l,mem))
            mem[i][j] += 1
            return mem[i][j]

        for i in range(m):
            for j in range(n):
                ans = max(ans,dfs(matrix,i,j,mem))
        return ans
        
```

```python
class Solution(object):
    def solveSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: None Do not return anything, modify board in-place instead.
        """
        
        def isValid(board,i,j,num):
            for k in range(9):
                if board[k][j] == num:
                    return False
                if board[i][k] == num:
                    return False
            start_i = (i//3)*3
            start_j = (j//3)*3
            for k in range(3):
                for l in range(3):
                    if board[start_i+k][start_j+l]==num:
                        return False
            return True
        
        def dfs(board,i,j):
            if i==9:return True
            if j>=9:return dfs(board,i+1,0)
            if board[i][j]=='.':
                for num in range(1,10):
                    if isValid(board,i,j,str(num)):
                        board[i][j] = str(num)
                        if dfs(board,i,j+1): 
                            return True
                        board[i][j] = '.'
            else:
                return dfs(board,i,j+1)
            return False
                
        dfs(board,0,0)
```

# 54. Spiral Matrix

```python
class Solution(object):
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        m = len(matrix)
        if m == 0: return []
        n = len(matrix[0])
        
        ans = []
        
        up= 0
        down = m-1
        left=0
        right = n-1
        
        while True: 
            for j in range(left,right+1):
                ans.append(matrix[up][j])
            up += 1
            if up>down: break
            for i in range(up,down+1):
                ans.append(matrix[i][right])
            right -= 1
            if right<left: break
            for j in range(right,left-1,-1):
                ans.append(matrix[down][j])
            down -= 1
            if down<up:break
            for i in range(down,up-1,-1):
                ans.append(matrix[i][left])
            left += 1
            if left>right: break
        return ans
```

```python
class Solution(object):
    def generateMatrix(self, n):
        """
        :type n: int
        :rtype: List[List[int]]
        """
        if n==0: return []
        ans = [[0]*n for _ in range(n)]
        up, down = 0, n-1
        left,right = 0, n-1
        num = 0
        while True:
            for j in range(left,right+1):
                num += 1
                ans[up][j] = num
            up += 1
            if up>down: break
            for i in range(up,down+1):
                num += 1
                ans[i][right] = num
            right -= 1
            if right<left: break
            for j in range(right,left-1,-1):
                num += 1
                ans[down][j] = num
            down -= 1
            if down<up:break
            for i in range(down,up-1,-1):
                num+=1
                ans[i][left] = num
            left += 1
            if left>right: break
        return ans
```

```python
class Solution(object):
    def shortestDistance(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        m,n = len(grid),len(grid[0])
        dst = [[0]*n for _ in range(m)]
        cnt = [[0]*n for _ in range(m)]
        directions = [[-1,0],[1,0],[0,-1],[0,1]]
        ans = float('inf')
        
        def bfs(grid,i,j,dst):
            m,n = len(grid),len(grid[0])
            
            queue = collections.deque()
            queue.append((i,j,0))
            visited = set()
            visited.add((i,j))
            while queue:
                i,j,step = queue.popleft()
                dst[i][j] += step
                cnt[i][j] += 1
                for d in directions:
                    x = i+d[0]
                    y = j+d[1]
                    if x<0 or x>=m or y<0 or y>=n:continue
                    if (x,y) not in visited and grid[x][y]==0:
                        queue.append((x,y,step+1))
                        visited.add((x,y))
        
        totCnt = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    totCnt += 1
                    bfs(grid,i,j,dst)
        
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 0 and cnt[i][j]==totCnt:
                    ans = min(ans, dst[i][j])
        if ans == float('inf'): return -1
        return ans
```

# 296. Best Meeting Point

```python
class Solution(object):
    def minTotalDistance(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        m,n = len(grid),len(grid[0])
        rows = []
        cols = []
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    rows.append(i)
                    cols.append(j)
        rows.sort()
        cols.sort()
        x = rows[len(rows)//2]
        y = cols[len(cols)//2]
        ans = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j]==1:
                    ans += abs(i-x)+abs(j-y)
        return ans
```

# 348. Design Tic-Tac-Toe

```python
class TicTacToe(object):

    def __init__(self, n):
        """
        Initialize your data structure here.
        :type n: int
        """
        self.rows = [0]*n
        self.cols = [0]*n
        self.dc1 = 0
        self.dc2 = 0
        self.n = n
        

    def move(self, row, col, player):
        """
        Player {player} makes a move at ({row}, {col}).
        @param row The row of the board.
        @param col The column of the board.
        @param player The player, can be either 1 or 2.
        @return The current winning condition, can be either:
                0: No one wins.
                1: Player 1 wins.
                2: Player 2 wins.
        :type row: int
        :type col: int
        :type player: int
        :rtype: int
        """
        if player == 1:
            val = 1
        else:
            val = -1
            
        self.rows[row] += val
        self.cols[col] += val
        if row == col:
            self.dc1 += val
        if row+col== self.n-1:
            self.dc2 += val
        
        n = self.n
        if abs(self.rows[row])==n or abs(self.cols[col]) == n or abs(self.dc1)==n or abs(self.dc2)==n:
            return player
        
        return 0
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def reorderList(self, head):
        """
        :type head: ListNode
        :rtype: None Do not return anything, modify head in-place instead.
        """
        if not head or not head.next: return head
        p1 = head
        p2 = head
        while p2.next and p2.next.next:
            p1 = p1.next
            p2 = p2.next.next
            
        tmp = p1.next
        p1.next = None
        p1 = tmp
        
        def reverse(head):
            if not head or not head.next: return head
            pre = head
            cur = head.next
            head.next = None
            while cur:
                tmp = cur.next
                cur.next = pre
                pre = cur
                cur = tmp
            return pre
        
        def merge(head1,head2):
            first = True
            dummy = ListNode(None)
            p = dummy
            while head1 and head2:
                if first:
                    p.next = head1
                    head1 = head1.next
                else:
                    p.next = head2
                    head2 = head2.next
                p = p.next
                first = not first
            if head1: p.next = head1
            if head2: p.next = head2
            return dummy.next
        p1 = reverse(p1)
        return merge(head,p1)
```

# 138. Copy List with Random Pointer

```python
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, next, random):
        self.val = val
        self.next = next
        self.random = random
"""
class Solution(object):
    def copyRandomList(self, head):
        """
        :type head: Node
        :rtype: Node
        """
        if not head: return head
        visited  = {}
        
        def dfs(p, visited):
            if not p: return None
            if p in visited:
                return visited[p]
            node = Node(p.val,None,None)
            visited[p] = node
            node.next = dfs(p.next,visited)
            node.random = dfs(p.random,visited)
            return node
            
        dfs(head,visited)
        return visited[head]
```

# 144. Binary Tree Preorder Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        ans = []
        stack = [root]
        while stack:
            node = stack.pop()
            ans.append(node.val)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
        return ans
```

# 314. Binary Tree Vertical Order Traversal

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def verticalOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        
        queue = collections.deque()
        queue.append((root,0))
        ans = []
        hmap = collections.defaultdict(list)
        minLevel = float('inf')
        maxLevel = -float('inf')
        while queue:
            node,level = queue.popleft()
            hmap[level].append(node.val)
            minLevel = min(minLevel,level)
            maxLevel = max(maxLevel,level)
            if node.left:
                queue.append((node.left,level-1))
            if node.right:
                queue.append((node.right,level+1))
        
        for level in range(minLevel,maxLevel+1):
            ans.append(hmap[level])
        return ans
```

# 114. Flatten Binary Tree to Linked List

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def flatten(self, root):
        """
        :type root: TreeNode
        :rtype: None Do not return anything, modify root in-place instead.
        """
        if not root: return None
        stack = []
        cur = root
        while cur or stack:
            if cur.right:
                stack.append(cur.right)
            if cur.left:
                cur.right = cur.left
                cur.left = None
            elif stack:
                node = stack.pop()
                cur.right = node
            cur = cur.right
        return root
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isBalanced(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        def helper(root):
            if not root: return (True, 0)
            leftCheck, leftD = helper(root.left)
            rightCheck, rightD = helper(root.right)
            if not leftCheck or not rightCheck:
                return (False, None)
            if abs(leftD-rightD)<=1:
                return (True, 1+max(leftD,rightD))
            else:
                return (False,None)
            
        ans,_ = helper(root)
        return ans
```

# 173. Binary Search Tree Iterator

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class BSTIterator(object):

    def __init__(self, root):
        """
        :type root: TreeNode
        """
        self.ans = collections.deque()
        def inorder(root):
            if root is None: return
            inorder(root.left)
            self.ans.append(root.val)
            inorder(root.right)
        inorder(root)

    def next(self):
        """
        @return the next smallest number
        :rtype: int
        """
        if self.hasNext():
            return self.ans.popleft()
        

    def hasNext(self):
        """
        @return whether we have a next smallest number
        :rtype: bool
        """
        return len(self.ans)>0
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def findFrequentTreeSum(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        self.hmap = collections.defaultdict(int)
        self.mostFreq = 0
        def helper(root):
            if root is None: return 0
            left = helper(root.left)
            right = helper(root.right)
            val = root.val + left+right
            self.hmap[val]+=1
            self.mostFreq = max(self.mostFreq,self.hmap[val])
            return val
        
        helper(root)
        ans = []
        for k, v in self.hmap.items():
            if v == self.mostFreq:
                ans.append(k)
        return ans
```

```python
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, left, right, next):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
"""
class Solution(object):
    def connect(self, root):
        """
        :type root: Node
        :rtype: Node
        """
        if not root: return root
        queue = collections.deque()
        queue.append(root)
        while queue:
            size = len(queue)
            prev = None
            for _ in range(size):
                node = queue.popleft()
                if prev:
                    prev.next = node
                prev = node
                if node.left:
                    queue.append(node.left)
                    queue.append(node.right)
        return root
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        
        def helper(root, string):
            if root is None: return string+'#,'
            string += str(root.val)
            string += ','
            string = helper(root.left, string)
            string = helper(root.right, string)
            return string
        return helper(root,'')
        
        

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        def helper(strList):
            if not strList: return
            val = strList.pop(0)
            if val == '#':
                return None
            node = TreeNode(int(val))
            node.left = helper(strList)
            node.right = helper(strList)
            return node
            
        data = data.rstrip(',')
        return helper(data.split(','))
```

# 333. Largest BST Subtree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def largestBSTSubtree(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        
        def helper(root):
            if root is None: return (True, 0, float('inf'), -float('inf'))
            lbst, lsize, lmin, lmax = helper(root.left)
            rbst, rsize, rmin, rmax = helper(root.right)
            if lbst and rbst and root.val > lmax and root.val < rmin:
                lmin = min(lmin, root.val)
                rmax = max(rmax, root.val)
                return (True, lsize+rsize+1, lmin, rmax)
            else:
                return (False, max(lsize, rsize), None, None)
            
        _, ans, _, _ = helper(root)
        return ans
```

# 208. Implement Trie (Prefix Tree)

```python
class Trie(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.data ={}

    def insert(self, word):
        """
        Inserts a word into the trie.
        :type word: str
        :rtype: None
        """
        p = self.data
        for w in word:
            p = p.setdefault(w,{})
        p['#'] = None 
        

    def search(self, word):
        """
        Returns if the word is in the trie.
        :type word: str
        :rtype: bool
        """
        p = self.data
        for w in word:
            if w in p:
                p = p[w]
            else:
                return False
        if '#' in p: return True
        return False
    
    def startsWith(self, prefix):
        """
        Returns if there is any word in the trie that starts with the given prefix.
        :type prefix: str
        :rtype: bool
        """
        p = self.data
        for w in prefix:
            if w in p:
                p = p[w]
            else:
                return False
        return True
```

# 211. Add and Search Word - Data structure design

```python
class WordDictionary(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.data = {}

    def addWord(self, word):
        """
        Adds a word into the data structure.
        :type word: str
        :rtype: None
        """
        p = self.data
        for w in word:
            p = p.setdefault(w,{})
        p['#'] = None
        

    def search(self, word):
        """
        Returns if the word is in the data structure. A word could contain the dot character '.' to represent any one letter.
        :type word: str
        :rtype: bool
        """
        def helper(p, word, i):
            if i>=len(word):
                return True if '#' in p else False
            c = word[i]
            if c!= '.':
                if c not in p: return False
                else: return helper(p[c],word,i+1)
            else:
                ans = False
                for k in p.keys():
                    if k=='#': continue
                    ans = (ans or helper(p[k],word,i+1))
                return ans
        
        return helper(self.data, word, 0)
```
