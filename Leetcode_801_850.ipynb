# 804. Unique Morse Code Words

```python
class Solution(object):
    def uniqueMorseRepresentations(self, words):
        """
        :type words: List[str]
        :rtype: int
        """
        letters = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q',\
                   'r','s','t','u','v','w','x','y','z']
        morses = [".-","-...","-.-.","-..",".","..-.","--.","....","..",".---","-.-",".-..",\
                 "--","-.","---",".--.","--.-",".-.","...","-","..-","...-",".--","-..-","-.--","--.."]
        letter_map = dict(zip(letters,morses))
        root = {}
        cnt = 0
        for word in words:
            p = root
            morse = ''
            for c in word:
                morse += letter_map[c.lower()]
            for m in morse:
                p = p.setdefault(m,{})
            if '_END' not in p:
                p['_END'] = None
                cnt +=1
        return cnt
```

# 806. Number of Lines To Write String

```python
class Solution(object):
    def numberOfLines(self, widths, S):
        """
        :type widths: List[int]
        :type S: str
        :rtype: List[int]
        """
        lines = 0
        width = 0
        for c in S:
            width += widths[ord(c)-ord('a')]
            if width > 100:
                lines+=1
                width = widths[ord(c)-ord('a')]
        return [lines+1,width]
        
```

# 811. Subdomain Visit Count

```python
class Solution(object):
    def subdomainVisits(self, cpdomains):
        """
        :type cpdomains: List[str]
        :rtype: List[str]
        """
        hashmap = {}
        for domain in cpdomains:
            cnt, name = domain.split(' ')
            cnt = int(cnt)
            name_list = name.split('.')
            for i in range(len(name_list)):
                sub_name = '.'.join(name_list[i:])
                if sub_name not in hashmap:
                    hashmap[sub_name] = cnt
                else:
                    hashmap[sub_name] += cnt
        ans = []
        for k,v in hashmap.items():
            ans.append(str(v)+" "+k)
        return ans
        
```

# 812. Largest Triangle Area

```python
class Solution(object):
    def largestTriangleArea(self, points):
        """
        :type points: List[List[int]]
        :rtype: float
        """
        def area(p, q, r):
            return .5 * abs(p[0]*q[1]+q[0]*r[1]+r[0]*p[1]
                           -p[1]*q[0]-q[1]*r[0]-r[1]*p[0])

        return max(area(*triangle)
            for triangle in itertools.combinations(points, 3))
        
```

# 821. Shortest Distance to a Character

```python
class Solution(object):
    def shortestToChar(self, S, C):
        """
        :type S: str
        :type C: str
        :rtype: List[int]
        """
        n = len(S)
        ans = [0] * n
        prev = -float('inf')
        for i in range(n):
            if S[i]==C: prev = i
            else:ans[i] = i-prev
            
        prev = float('inf')
        for i in range(n-1,-1,-1):
            if S[i] ==C: prev = i
            else: ans[i] = min(prev -i,ans[i])
        return ans
                
            
```

# 824. Goat Latin

```python
class Solution(object):
    def toGoatLatin(self, S):
        """
        :type S: str
        :rtype: str
        """
        ans = []
        words = S.split(' ')
        cnt_a = 0
        for word in words:
            cnt_a += 1
            if word[0].lower() in ['a','e','i','o','u']:
                ans.append(word+'ma'+'a'*cnt_a)
            else:
                ans.append(word[1:]+word[0]+'ma'+'a'*cnt_a)
        return ' '.join(ans)
            
            
```

# 830. Positions of Large Groups

```python
class Solution(object):
    def largeGroupPositions(self, S):
        """
        :type S: str
        :rtype: List[List[int]]
        """
        n = len(S)
        anchor = 0
        cnt = 0
        ans =[]
        for i in range(n): 
            if S[i] == S[anchor]:
                cnt+= 1
            else:
                if cnt >= 3:
                    ans.append([anchor,i-1])
                anchor = i
                cnt = 1
        if cnt >= 3:
            ans.append([anchor,n-1])
        return ans
```

# 832. Flipping an Image

```python
class Solution(object):
    def flipAndInvertImage(self, A):
        """
        :type A: List[List[int]]
        :rtype: List[List[int]]
        """
        m,n = len(A),len(A[0])
        for i in range(m):
            for j in range(n//2):
                A[i][j],A[i][n-1-j] = 1-A[i][n-1-j],1-A[i][j]
            if n%2!=0:
                A[i][n//2] = 1-A[i][n//2]
        return A
```

# 836. Rectangle Overlap

```python
class Solution(object):
    def isRectangleOverlap(self, rec1, rec2):
        """
        :type rec1: List[int]
        :type rec2: List[int]
        :rtype: bool
        """
        rec1_left = rec1[0]
        rec1_right = rec1[2]
        rec1_top = rec1[3]
        rec1_bottom = rec1[1]
        
        rec2_left = rec2[0]
        rec2_right = rec2[2]
        rec2_top = rec2[3]
        rec2_bottom = rec2[1]
        
        if min(rec1_right,rec2_right)> max(rec1_left,rec2_left) and \
        min(rec1_top,rec2_top)> max(rec1_bottom,rec2_bottom):
            return True
        return False
```

# 840. Magic Squares In Grid

```python
class Solution(object):
    def numMagicSquaresInside(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        R,C = len(grid),len(grid[0])
        cnt = 0
        for i in range(R-2):
            for j in range(C-2):
                temp = [row[j:j+3] for row in grid[i:i+3]]
                if temp[1][1]==5:
                    if temp[0][0] in [2,4,6,8] and temp[2][2]+temp[0][0]==10:
                        if temp[0][2] in [2,4,6,8] and temp[0][2]!=temp[0][0] and \
                        temp[0][2]!=temp[2][2] and temp[0][2]+temp[2][0]==10:
                            if sum([temp[0][k] for k in range(3)])==15 and sum([temp[2][k] for k in range(3)])==15 and\
                            sum([temp[k][0] for k in range(3)])==15 and sum([temp[2][k] for k in range(3)])==15:
                                cnt += 1
        return cnt
                
```

# 844. Backspace String Compare

```python
class Solution(object):
    def backspaceCompare(self, S, T):
        """
        :type S: str
        :type T: str
        :rtype: bool
        """
        s = []
        t = []
        for c in S:
            if c=='#':
                if s:s.pop()
            else:s.append(c)
        for c in T:
            if c=='#':
                if t:t.pop()
            else:t.append(c)
    
        if ''.join(s) ==''.join(t):
            return True
        return False
```
