# 703. Kth Largest Element in a Stream

```python
import heapq
class KthLargest(object):

    def __init__(self, k, nums):
        """
        :type k: int
        :type nums: List[int]
        """
        self.k = k
        self.k_list = heapq.nlargest(k, nums)
        heapq.heapify(self.k_list)

    def add(self, val):
        """
        :type val: int
        :rtype: int
        """
        if self.k_list.__len__() < self.k:
            heapq.heappush(self.k_list, val)
        else:
            heapq.heappushpop(self.k_list, val)
        return self.k_list[0]
```

# 704. Binary Search

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        n = len(nums)
        i = 0
        j = n-1
        while i<=j:
            mid = (i+j)//2
            if nums[mid] == target: return mid
            elif nums[mid]< target: i = mid+1
            else: j = mid -1
        return -1
```

# 705. Design HashSet

```python
class MyHashSet(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.data = set()
        

    def add(self, key):
        """
        :type key: int
        :rtype: void
        """
        self.data.add(key)
        

    def remove(self, key):
        """
        :type key: int
        :rtype: void
        """
        if key in self.data: self.data.remove(key)

    def contains(self, key):
        """
        Returns true if this set contains the specified element
        :type key: int
        :rtype: bool
        """
        return key in self.data
```

# 706. Design HashMap

```python
class MyHashMap(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.data = {}

    def put(self, key, value):
        """
        value will always be non-negative.
        :type key: int
        :type value: int
        :rtype: void
        """
        self.data[key] = value

    def get(self, key):
        """
        Returns the value to which the specified key is mapped, or -1 if this map contains no mapping for the key
        :type key: int
        :rtype: int
        """
        if key in self.data: return self.data[key]
        else: return -1
        

    def remove(self, key):
        """
        Removes the mapping of the specified value key if this map contains a mapping for the key
        :type key: int
        :rtype: void
        """
        if key in self.data: self.data.pop(key)
    
```

# 707. Design Linked List

```python
class MyLinkedList(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.data = []
        

    def get(self, index):
        """
        Get the value of the index-th node in the linked list. If the index is invalid, return -1.
        :type index: int
        :rtype: int
        """
        if 0<=index<len(self.data):
            return self.data[index]
        else:
            return -1
        

    def addAtHead(self, val):
        """
        Add a node of value val before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
        :type val: int
        :rtype: void
        """
        self.data.insert(0,val)
        

    def addAtTail(self, val):
        """
        Append a node of value val to the last element of the linked list.
        :type val: int
        :rtype: void
        """
        self.data.append(val)
        

    def addAtIndex(self, index, val):
        """
        Add a node of value val before the index-th node in the linked list. If index equals to the length of linked list, the node will be appended to the end of linked list. If index is greater than the length, the node will not be inserted.
        :type index: int
        :type val: int
        :rtype: void
        """
        if 0<=index<=len(self.data):
            self.data.insert(index,val)
        
        
        

    def deleteAtIndex(self, index):
        """
        Delete the index-th node in the linked list, if the index is valid.
        :type index: int
        :rtype: void
        """
        if 0<=index<len(self.data):
            self.data.pop(index)
        
```

# 717. 1-bit and 2-bit Characters

```python
class Solution(object):
    def isOneBitCharacter(self, bits):
        """
        :type bits: List[int]
        :rtype: bool
        """
        i =0
        if bits[-1] != 0: return False
        while i < len(bits) -1:
            if bits[i] ==1:
                i += 2
            elif bits[i] == 0:
                i += 1
        return i == len(bits) -1
        
```

# 720. Longest Word in Dictionary

```python
class Solution(object):
    def longestWord(self, words):
        """
        :type words: List[str]
        :rtype: str
        """
        root = {}
        for i, word in enumerate(words):
            p = root
            cnt = 0
            for letter in word:
                p = p.setdefault(letter,{})
            p['_END_'] = i
        
        ans = ""
        stack = root.values()
        while stack:
            cur = stack.pop()
            if '_END_' in cur:
                word = words[cur['_END_']]
                if len(word)> len(ans) or len(word)==len(ans) and word<ans: ans = word
                for letter in cur:
                    if letter!='_END_':
                        stack.append(cur[letter])
        return ans
        
        
```

# 728. Self Dividing Numbers

```python
class Solution(object):
    def selfDividingNumbers(self, left, right):
        """
        :type left: int
        :type right: int
        :rtype: List[int]
        """
        def seldDividing(n):
            n1 = n
            while n1:
                d = n1%10
                if d==0 or n%d != 0: return False
                n1 = n1//10
            return True
        ans = []
        for n in range(left,right+1):
            if seldDividing(n): ans.append(n)
        return ans
                
```

# 733. Flood Fill

```python
class Solution(object):
    def floodFill(self, image, sr, sc, newColor):
        """
        :type image: List[List[int]]
        :type sr: int
        :type sc: int
        :type newColor: int
        :rtype: List[List[int]]
        """
        m,n = len(image), len(image[0])
        oldColor = image[sr][sc] 
        if oldColor == newColor: return image
        def helper(i,j):
            if image[i][j] != oldColor: return image
            image[i][j] = newColor
            if i>=1: helper(i-1,j)
            if i+1<m:helper(i+1,j)
            if j>=1: helper(i,j-1)
            if j+1<n: helper(i,j+1)
        helper(sr,sc)
        
        return image
    
    
```

# 744. Find Smallest Letter Greater Than Target

```python
class Solution(object):
    def nextGreatestLetter(self, letters, target):
        """
        :type letters: List[str]
        :type target: str
        :rtype: str
        """
        n = len(letters)
        low = 0
        high = n-1
        while low<=high:
            mid = (low+high)//2
            if letters[mid]>target:
                high = mid -1
            else:
                low = mid + 1
        return letters[low%n]
```

# 746. Min Cost Climbing Stairs

```python
class Solution(object):
    def minCostClimbingStairs(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """
        n = len(cost)
        if n<=1: return 0
        pay = [0]*(len(cost)+1)
        for i in range(2,len(cost)+1):
            pay[i] = min(pay[i-1]+cost[i-1],pay[i-2]+cost[i-2])
        return pay[-1]
            
```

# 747. Largest Number At Least Twice of Others

```python
import sys
class Solution(object):
    def dominantIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums)<=1:return -1
        max1 = -sys.maxsize
        max2 = -sys.maxsize
        largest = -1
        for i, n in enumerate(nums):
            if n> max1:
                max1,max2 = n, max1
                largest = i
            elif n> max2:
                max2 = n
        if max1>=max2*2:
            return largest
        else:
            return -1
```

# 748. Shortest Completing Word

```python
from collections import Counter
import sys
class Solution(object):
    def shortestCompletingWord(self, licensePlate, words):
        """
        :type licensePlate: str
        :type words: List[str]
        :rtype: str
        """
        license = [c.lower() for c in licensePlate if c.isalpha()]
        lic_cnt = Counter(license)
        shortest = sys.maxsize
        ans = ""
        for word in words:
            cnt = Counter(word)
            match = True
            for l in lic_cnt:
                if l not in cnt or cnt[l]<lic_cnt[l]:
                    match = False
            if match and len(word)<shortest:
                ans = word
                shortest = len(word)
        return ans
                
```
