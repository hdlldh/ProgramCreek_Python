# 905. Sort Array By Parity

```python
from collections import deque
class Solution(object):
    def sortArrayByParity(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        dq = deque()
        for n in A:
            if n%2==0:
                dq.appendleft(n)
            else:
                dq.append(n)
        return list(dq)
        
```

# 908. Smallest Range I

```python
class Solution(object):
    def smallestRangeI(self, A, K):
        """
        :type A: List[int]
        :type K: int
        :rtype: int
        """
        return max(0, max(A)-min(A)-2*K)
```

# 914. X of a Kind in a Deck of Cards

```python
from collections import Counter
class Solution(object):
    def hasGroupsSizeX(self, deck):
        """
        :type deck: List[int]
        :rtype: bool
        """
        counter = Counter(deck)
        cnt_vals = counter.values()
        def gcd(x,y):
            if x==0:
                return y
            else:
                return gcd(y%x,x)
        
        return reduce(gcd,cnt_vals)>=2
    
```

# 922. Sort Array By Parity II

```python
class Solution(object):
    def sortArrayByParityII(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        q1 = []
        q2 = []
        for i in range(len(A)):
            if i%2==0 and A[i]%2!=0:
                q1.append(A[i])
            if i%2!=0 and A[i]%2==0:
                q2.append(A[i])
        
        for i in range(len(A)):
            if i%2==0 and A[i]%2!=0:
                A[i] = q2.pop()
            if i%2!=0 and A[i]%2==0:
                A[i] = q1.pop()
        return A
                
            
        
```

# 925. Long Pressed Name

```python
class Solution(object):
    def isLongPressedName(self, name, typed):
        """
        :type name: str
        :type typed: str
        :rtype: bool
        """
        ns=nf=0
        ts=tf=0
        nsize = len(name)
        tsize = len(typed)
        if nsize>tsize: return False
        while nf<nsize and tf<tsize:
            while nf!=nsize-1 and name[nf]==name[nf+1]:
                nf +=1
            ncnt = nf-ns+1
            nletter = name[nf]
            while tf!=tsize-1 and typed[tf]==typed[tf+1]:
                tf +=1
            tcnt = tf-ts+1
            tletter = typed[tf]
            if nletter!=tletter or tcnt<ncnt:
                return False
            nf = nf+1
            ns = nf
            tf = tf+1
            ts = tf
        if nf!=nsize or tf!=tsize:return False
        return True 
        
       
```

# 929. Unique Email Addresses

```python
class Solution(object):
    def numUniqueEmails(self, emails):
        """
        :type emails: List[str]
        :rtype: int
        """
        ans = set()
        for em in emails:
            tmp = em.split('@')
            tmp[0]=tmp[0].replace('.','')
            if '+' in tmp[0]:
                tmp[0] = tmp[0][:tmp[0].index('+')]
            ans.add('@'.join(tmp))
        return len(ans)
```

# 933. Number of Recent Calls

```python
from collections import deque
class RecentCounter(object):

    def __init__(self):
        self.data = deque()
        

    def ping(self, t):
        """
        :type t: int
        :rtype: int
        """
        self.data.append(t)
        while self.data[0]<t-3000:
            self.data.popleft()
        return len(self.data)
```

# 937. Reorder Log Files

```python
class Solution(object):
    def reorderLogFiles(self, logs):
        """
        :type logs: List[str]
        :rtype: List[str]
        """
        def fun(log):
            one,rest = log.split(' ',1)
            if rest[0].isalpha():
                return (0, rest, one)
            else:
                return (1,)
        return sorted(logs,key=fun)
```

# 941. Valid Mountain Array

```python
class Solution(object):
    def validMountainArray(self, A):
        """
        :type A: List[int]
        :rtype: bool
        """
        n = len(A)
        if n<=2: return False
        inc = True
        for i in range(1,len(A)):
            if inc and A[i]==A[i-1]:return False
            if inc and A[i]<A[i-1]:
                if i==1: return False
                inc = False
            elif not inc and A[i]>=A[i-1]:
                return False
        if inc: return False
        return True
                
                
```

# 942. DI String Match

```python
class Solution(object):
    def diStringMatch(self, S):
        """
        :type S: str
        :rtype: List[int]
        """
        lo, hi = 0, len(S)
        ans = []
        for s in S:
            if s == 'I':
                ans.append(lo)
                lo += 1
            else:
                ans.append(hi)
                hi -= 1
        return ans+[lo]
```

# 944. Delete Columns to Make Sorted

```python
class Solution(object):
    def minDeletionSize(self, A):
        """
        :type A: List[str]
        :rtype: int
        """
        R, C = len(A),len(A[0])
        if R == 1: return 0
        ans = [0]*C
        for i in range(C):
            for j in range(1,R):
                if A[j][i]<A[j-1][i]:
                    ans[i] = 1
                    break
        return sum(ans)
```

# 949. Largest Time for Given Digits

```python
class Solution(object):
    def largestTimeFromDigits(self, A):
        """
        :type A: List[int]
        :rtype: str
        """
        ans = -1
        for h1, h2, m1, m2 in itertools.permutations(A):
            hours = 10 * h1 + h2
            mins = 10 * m1 + m2
            time = 60 * hours + mins
            if 0 <= hours < 24 and 0 <= mins < 60 and time > ans:
                ans = time
            
        hours = ans//60
        mins = ans %60
        return '%02d:%02d'%(hours,mins)
        
        
            
```
