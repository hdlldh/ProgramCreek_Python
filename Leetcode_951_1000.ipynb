# 953. Verifying an Alien Dictionary

```python
class Solution(object):
    def isAlienSorted(self, words, order):
        """
        :type words: List[str]
        :type order: str
        :rtype: bool
        """
        self.dmap = {}
        for i in range(len(order)):
            self.dmap[order[i]] = i
        def fun(word):
            return tuple([self.dmap[w] for w in word])
        sorted_words = sorted(words,key=fun)
        for n in range(len(words)):
            if sorted_words[n]!=words[n]:
                return False
        return True
```

# 961. N-Repeated Element in Size 2N Array

```python
class Solution(object):
    def repeatedNTimes(self, A):
        """
        :type A: List[int]
        :rtype: int
        """
        n = len(A)
        hashset = set()
        for i in range(n/2):
            if A[i] in hashset: return A[i]
            hashset.add(A[i])
```

# 965. Univalued Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isUnivalTree(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root: return True
        if root.left and root.left.val != root.val: return False
        if root.right and root.right.val != root.val: return False
        return self.isUnivalTree(root.left) and self.isUnivalTree(root.right)
```

# 970. Powerful Integers

```python
class Solution(object):
    def powerfulIntegers(self, x, y, bound):
        """
        :type x: int
        :type y: int
        :type bound: int
        :rtype: List[int]
        """
        ans = set()
        
        for i in range(18):
            for j in range(18):
                v = x**i +y**j
                if v<= bound:
                    ans.add(v)
        return list(ans)
```

# 973. K Closest Points to Origin

```python
from heapq import heappush, heappop
class Solution(object):
    def kClosest(self, points, K):
        """
        :type points: List[List[int]]
        :type K: int
        :rtype: List[List[int]]
        """
        plist = []
        for point in points:
            dist = point[0]**2 + point[1]**2
            heappush(plist,(dist,point))
        topk = []
        for _ in range(K):
            dist, point = heappop()
            topk.appen(point)
        return topk
```

# 976. Largest Perimeter Triangle

```python
class Solution(object):
    def largestPerimeter(self, A):
        """
        :type A: List[int]
        :rtype: int
        """
        A.sort()
        for i in range(len(A)-1,1,-1):
            c = A[i]
            b = A[i-1]
            a = A[i-2]
            if (a+b)>c: return a+b+c
        return 0
```

# 977. Squares of a Sorted Array

```python
class Solution(object):
    def sortedSquares(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        head = 0 
        tail = len(A) -1
        output = []
        while head<=tail:
            if abs(A[head]) >= abs(A[tail]):
                output.append(A[head]**2)
                head += 1
            else:
                output.append(A[tail]**2)
                tail -= 1
        output.reverse()
        return output
            
```

# 984. String Without AAA or BBB

```python
class Solution(object):
    def strWithout3a3b(self, A, B):
        """
        :type A: int
        :type B: int
        :rtype: str
        """
        
        countA = 0
        countB = 0
        ans = ''
        while A or B:
            if A >= B:
                if countA<2:
                    writeA = True
                else:
                    writeA = False
            else:
                if countB <2:
                    writeA = False
                else:
                    writeA = True
            if writeA:
                ans +='a'
                A -= 1
                countA += 1
                countB = 0
            else:
                ans +='b'
                B -= 1
                countB += 1
                countA = 0
                
        return ans
```

```python
class Solution(object):
    def sumEvenAfterQueries(self, A, queries):
        """
        :type A: List[int]
        :type queries: List[List[int]]
        :rtype: List[int]
        """
        hashmap = {}
        sum = 0
        ans = []
        for n in A:
            if n%2 == 0: sum += n
        
        for x, k in queries:
            if A[k]%2 ==0: sum -= A[k]
            A[k] += x
            if A[k]%2 ==0: sum += A[k]
                
            ans.append(sum)
        return ans
                
```

# 989. Add to Array-Form of Integer

```python
class Solution(object):
    def addToArrayForm(self, A, K):
        """
        :type A: List[int]
        :type K: int
        :rtype: List[int]
        """
        lenA = len(A)
        A.reverse()
        B = []
        while K:
            B.append(K%10)
            K = K//10
        lenB = len(B)
        c = 0
        C = [0]*max(lenA,lenB)
        for i in range(len(C)):
            a = A[i] if i<lenA else 0
            b = B[i] if i<lenB else 0
            C[i] = (a+b+c)%10
            c = 1 if a+b+c>=10 else 0
        if c: C.append(1)
        C.reverse()
        return C
```

# 993. Cousins in Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isCousins(self, root, x, y):
        """
        :type root: TreeNode
        :type x: int
        :type y: int
        :rtype: bool
        """
        parent = {}
        depth = {}
        def dfs(node,par=None):
            if node:
                depth[node.val] = 1+depth[par.val] if par else 0
                parent[node.val] = par
                dfs(node.left,node)
                dfs(node.right,node)
        dfs(root)
        return parent[x]!=parent[y] and depth[x]==depth[y]
```
