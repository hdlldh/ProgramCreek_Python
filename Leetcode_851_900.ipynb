# 852. Peak Index in a Mountain Array

```python
class Solution(object):
    def peakIndexInMountainArray(self, A):
        """
        :type A: List[int]
        :rtype: int
        """
        n = len(A)
        low = 0
        high = n-1
        while low<=high:
            mid = (low+high)//2
            if (mid==0 or A[mid]>A[mid-1]) and (mid==n-1 or A[mid]>A[mid+1]):
                return mid
            elif A[mid] < A[mid-1] and A[mid]>A[mid+1]:
                high = mid -1
            else:
                low = mid+1
        return low
```

# 859. Buddy Strings

```python
class Solution(object):
    def buddyStrings(self, A, B):
        """
        :type A: str
        :type B: str
        :rtype: bool
        """
        n, m = len(A),len(B)
        if n!=m: return False
        
        mismatch = []
        for i in range(n):
            if A[i]!=B[i]:
                mismatch.append((A[i],B[i]))
                
        if not mismatch: return len(set(A))!=len(A)
        if len(mismatch)!=2: return False
        return mismatch[0][0]==mismatch[1][1] and  mismatch[0][1]==mismatch[1][0]
```

# 860. Lemonade Change

```python
from collections import deque
class Solution(object):
    def lemonadeChange(self, bills):
        """
        :type bills: List[int]
        :rtype: bool
        """
        dq = deque()
        for b in bills:
            print(dq)
            if b==5:
                dq.appendleft(5)
            elif b==10:
                if not dq or dq[0]!=5: return False
                dq.popleft()
                dq.append(10)
            else:
                if not dq: return False
                if dq[-1]==10 and dq[0]==5:
                    dq.popleft()
                    dq.pop()
                elif dq[-1]==5 and len(dq)>=3:
                    dq.pop()
                    dq.pop()
                    dq.pop()
                else:
                    return False
                    
        return True
                
                
```

# 867. Transpose Matrix

```python
class Solution(object):
    def transpose(self, A):
        """
        :type A: List[List[int]]
        :rtype: List[List[int]]
        """
        R,C = len(A),len(A[0])
        B =[[0]*R for _ in range(C)]
        for i in range(R):
            for j in range(C):
                B[j][i]=A[i][j]
        return B
```

# 868. Binary Gap

```python
class Solution(object):
    def binaryGap(self, N):
        """
        :type N: int
        :rtype: int
        """
        prev = None
        maxGap = 0
        cnt = 0
        while N:
            if N%2 == 1:
                if prev is not None: maxGap = max(maxGap,cnt-prev)
                prev = cnt   
            cnt+=1
            N = N//2
        return maxGap  
```

# 872. Leaf-Similar Trees

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def leafSimilar(self, root1, root2):
        """
        :type root1: TreeNode
        :type root2: TreeNode
        :rtype: bool
        """
        self.tmp = []
        def inorder(root):
            if not root: return
            inorder(root.left)
            if not root.left and not root.right:
                self.tmp.append(root.val)
            inorder(root.right)
        inorder(root1)
        leaves1 = [leaf for leaf in self.tmp]
        self.tmp = []
        inorder(root2)
        leaves2 = [leaf for leaf in self.tmp]
        if len(leaves1)!=len(leaves2):
            return False
        for i in range(len(leaves1)):
            if leaves1[i]!=leaves2[i]:
                return False
        return True
        
```

# 874. Walking Robot Simulation

```python
class Solution(object):
    def robotSim(self, commands, obstacles):
        """
        :type commands: List[int]
        :type obstacles: List[List[int]]
        :rtype: int
        """
        dx = [0,1,0,-1]
        dy = [1,0,-1,0]
        
        x = y = di = 0
        obstacleSet = set()
        for ob in obstacles:
            obstacleSet.add(tuple(ob))
        ans = 0
        
        for cmd in commands:
            if cmd == -2:
                di = (di-1)%4
            elif cmd == -1:
                di = (di+1)%4
            else:
                for k in range(cmd):
                    if (x+dx[di], y+ dy[di]) not in obstacleSet:
                        x += dx[di]
                        y += dy[di]
                        ans = max(ans, x**2+y**2)
        return ans
```

# 884. Uncommon Words from Two Sentences

```python
from collections import Counter
class Solution(object):
    def uncommonFromSentences(self, A, B):
        """
        :type A: str
        :type B: str
        :rtype: List[str]
        """
        counter_A = Counter(A.split(' '))
        counter_B = Counter(B.split(' '))
        ans = []
        for key,val in counter_A.items():
            if val==1 and key not in counter_B:
                ans.append(key)
        
        for key,val in counter_B.items():
            if val==1 and key not in counter_A:
                ans.append(key)
        return ans
```

# 876. Middle of the Linked List

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def middleNode(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head or not head.next: return head
        slow = head
        fast = head
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next
        return slow
```

# 883. Projection Area of 3D Shapes

```python
class Solution(object):
    def projectionArea(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        ans = 0
        R, C = len(grid), len(grid[0])
        xz = [0]*C
        yz = [0]*R
        xy = 0
        for i in range(R):
            for j in range(C):
                if grid[i][j]> 0:
                    xy +=1
                xz[j] = max(xz[j],grid[i][j])
                yz[i] = max(yz[i],grid[i][j])
        return xy+sum(xz)+sum(yz)
        
```

# 888. Fair Candy Swap

```python
class Solution(object):
    def fairCandySwap(self, A, B):
        """
        :type A: List[int]
        :type B: List[int]
        :rtype: List[int]
        """
        setB= set(B)
        sumA = sum(A)
        sumB = sum(B)
        for x in A:
            y = (sumB-sumA)//2 +x
            if y in setB:
                return [x,y]
        return []
```

# 892. Surface Area of 3D Shapes

```python
class Solution(object):
    def surfaceArea(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        N = len(grid)
        ans = 0
        for x in range(N):
            for y in range(N):
                if grid[x][y] >0:
                    ans += 2
                    for nx,ny in [(x+1,y),(x-1,y),(x,y+1),(x,y-1)]:
                        if 0<=nx<N and 0<=ny<N:
                            nval = grid[nx][ny]
                        else:
                            nval = 0
                        ans += max(grid[x][y]-nval,0)
        return ans
```

# 893. Groups of Special-Equivalent Strings

```python
class Solution(object):
    def numSpecialEquivGroups(self, A):
        """
        :type A: List[str]
        :rtype: int
        """
        B = set()
        for a in A:
            sig = ''.join(sorted(a[::2]))+''.join(sorted(a[1::2]))
            B.add(sig)
        return len(B)
```

# 896. Monotonic Array

```python
class Solution(object):
    def isMonotonic(self, A):
        """
        :type A: List[int]
        :rtype: bool
        """
        n = len(A)
        dec = inc = True
        for i in range(1,len(A)):
            if A[i] > A[i-1]:
                dec = False
            if A[i] < A[i-1]:
                inc = False
        return inc or dec
```

# 897. Increasing Order Search Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def increasingBST(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        
        ans = self.cur = TreeNode(None)
        return ans.right
    
    def inorder(self,node):
        if node:
            self.inorder(node.left)
            new_node = TreeNode(node.val)
            self.cur.right = new_node
            self.cur = self.cur.right
            self.inorder(node.right)
```
